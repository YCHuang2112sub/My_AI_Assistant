```python
import sounddevice as sd
import numpy as np

def record_audio(duration, filename):
    print("Recording...")
    audio_data = sd.rec(int(duration * 44100), samplerate=44100, channels=1, dtype='int16')
    sd.wait()
    print("Finished recording.")
    sd.play(audio_data, 44100)
    sd.wait()


    with open(filename, 'wb') as file:
        np.save(file, audio_data)

def play_audio(filename):
    print("Playing...")
    with open(filename, 'rb') as file:
        audio_data = np.load(file)
        sd.play(audio_data, 44100)
        sd.wait()
    print("Finished playing.")

if __name__ == "__main__":
    duration = 5  # Adjust this as needed
    output_file = "output.npy"

    record_audio(duration, output_file)
    play_audio(output_file)
```
